```python
import csv
import pandas as pd
import numpy as np
import re
import spacy
```

```python
col_list=["reviews.text","brand"]
full = pd.read_csv("245_1.csv", usecols=col_list)
print(len(full))
```

```python
train = full.head(53000)
print(len(train["brand"]))
test = full.tail(71044-53000)
print(len(test['brand']))
```

```python
brands = []
brands = full['brand'].unique().astype(str)
print(type(brands))
brands = brands.tolist()
print(len(brands))
```

```python
train_brands = []
train_brands = train['brand'].unique().astype(str)
print(type(train_brands))
train_brands = train_brands.tolist()
print(len(train_brands))
train_brands.remove('No Brand')
print(len(train_brands))
```

```python
test_brands = []
test_brands = test['brand'].unique().astype(str)
print(type(test_brands))
test_brands = test_brands.tolist()
print(len(test_brands))
```

```python
print(test["brand"][53000])
```

```python
common = []
uncommon=[]
for i in train_brands:
    if i in test_brands:
        common.append(i)
for i in test_brands:    
    if i not in train_brands:
        uncommon.append(i)
print(len(common))
print(len(uncommon))
```

```python
validation = []
reviews=[]
#a= "wipes"
for j in range(len(common)):
    b = common[j]
    print(j)
    for i in range(int(len(test))):
        if test["brand"][i+53000]==b:
            Sent = test["reviews.text"][i+53000]
            if str(b).lower() in str(Sent).lower():
                match = re.search(b, Sent)
                if match != None:
                    myDict = {}
                    myDict["entities"] = [(match.start(),match.end(), "PRODUCT")]
                    validation.append((Sent,myDict))
    print(len(validation))
    print(b)
```

```python
training = []
reviews=[]
#a= "wipes"
for j in range(len(train_brands)):
    a = train_brands[j]
    print(j)
    for i in range(int(len(train))):
        if train["brand"][i]==a:
            Sent = train["reviews.text"][i]
            if str(a).lower() in str(Sent).lower():
                match = re.search(a, Sent)
                if match != None:
                    myDict = {}
                    myDict["entities"] = [(match.start(),match.end(), "PRODUCT")]
                    training.append((Sent,myDict))
    print(len(training))
    print(a)
```

```python
for i in range(len(training)):
    if i==0:
        f = open("training.txt", "w")
    else:
        f = open("training.txt", "a")
    f.write(str(training[i]))
    if i != (len(training)-1):
        f.write("\n")
    f.close()
len(training)
```

```python
for i in range(len(validation)):
    if i==0:
        f = open("validation.txt", "w")
    else:
        f = open("validation.txt", "a")
    f.write(str(validation[i]))
    if i != (len(validation)-1):
        f.write("\n")
    f.close()
len(validation)
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
ner=nlp.get_pipe("ner")
```

```python
for _, annotations in training:
    for ent in annotations.get("entities"):
        ner.add_label(ent[2])
```

```python
pipe_exceptions = ["ner", "trf_wordpiecer", "trf_tok2vec"]
unaffected_pipes = [pipe for pipe in nlp.pipe_names if pipe not in pipe_exceptions]
```

```python
import random
from spacy.util import minibatch, compounding
from pathlib import Path

# TRAINING THE MODEL
with nlp.disable_pipes(*unaffected_pipes):

  # Training for 30 iterations
  for iteration in range(30):

    # shuufling examples  before every iteration
    random.shuffle(training)
    losses = {}
    # batch up the examples using spaCy's minibatch
    batches = minibatch(training, size=compounding(4.0, 32.0, 1.001))
    for batch in batches:
        texts, annotations = zip(*batch)
        nlp.update(
                    texts,  # batch of texts
                    annotations,  # batch of annotations
                    drop=0.5,  # dropout - make it harder to memorise data
                    losses=losses,
                )
        print("Losses", losses)
```
